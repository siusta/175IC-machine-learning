```
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd

data = {
    'size':['S','M',np.nan,'L','M','L','XL','M',np.nan,'S','XL','L',np.nan,'XL','M','S'],
    'color':['green','black',np.nan,'purple','pink','orange','yellow','brown',np.nan,'cyan',np.nan,'blue','green','black','red','purple'],
    'gender':['female',np.nan,'male',np.nan,'female','female','male',np.nan,'female','male','male','female','female','female',np.nan,'male'],
    'price':[119.0, 89.0, 84.0, 56.0, np.nan, 249.0, 29.99, 45.0, 99.0, 189.0, 39.90, np.nan, 56.0, 112.0,np.nan, 189.0],
    'weight':[380.0, 410.0, np.nan, 550.0, 320.0, np.nan, 679.0, 410.0, 520.0, np.nan, 415.0, 470.0, np.nan, 550.0, np.nan, 333.0],
    'bought':['no', 'yes', 'yes', 'no', np.nan, 'no', 'yes', 'no', 'yes', 'no', 'no', 'yes', 'no', 'yes', 'no', 'yes']
}
df_raw = pd.DataFrame(data=data)
df = df_raw.copy()

#1
df.dropna(thresh=len(df.keys())-2, inplace=True)

#2
df['price'].fillna(df['price'].mean(), inplace=True)
#3
df['weight'].fillna(df['weight'].interpolate(), inplace=True)

#4
df['size'].fillna('no size', inplace=True)
df['color'].fillna('multicolor', inplace=True)
df['gender'].fillna('unisex', inplace=True)

#5
df['cost'] = pd.cut(x=df['price'], bins=3, labels=('cheap', 'normal_price', 'expensive'))

#6
df['bought_yes'] = df['bought'].apply(lambda x: 1 if 'yes' == x else 0)

#7
input=df[['size','color','gender','price','cost','weight']]
output=df['bought_yes']
X_train, X_test, y_train, y_test = train_test_split(input,output, train_size=0.8)
print(X_train)
print(y_train)
```
