```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import plotly.express as px
from sklearn.ensemble import IsolationForest
```

```
np.random.seed(10)
data= make_blobs(n_samples=1250, centers=2, cluster_std=2.0, random_state=26)[0]

outlier = IsolationForest(n_estimators=30, contamination=0.26)
outlier.fit(data)

y_pred = outlier.predict(data)

all_data = np.c_[data, y_pred]
exception = all_data[all_data[:,2] == -1]
matching = all_data[all_data[:,2] == 1]


plt.figure(figsize=(12,7))
plt.scatter(matching[:,0],matching[:,1],label='matching',cmap='tab10')
plt.scatter(exception[:,0],exception[:,1],label='exceptions',cmap='tab10')
legend = plt.legend()
legend.legendHandles[1]._sizes = [40]

plt.show()
```
