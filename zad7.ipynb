```
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
```

```
data, _ = make_blobs(n_samples=1000, centers=4, random_state=15, cluster_std=1.25)
df = pd.DataFrame(data, columns=['x1','x2'])
px.scatter(df, 'x1','x2')
```

```
cluster = DBSCAN(eps=0.85, min_samples=24)
cluster.fit(data)
df['cluster'] = cluster.labels_
px.scatter(df,'x1','x2','cluster',color_continuous_midpoint=0)
```

```
df['cluster'].value_counts()
```

```
kmeans = KMeans(n_clusters=4, random_state=27).fit(data)
df['cluster']=kmeans.labels_
px.scatter(df,'x1','x2','cluster',color_continuous_midpoint=0)
```

```
df['kmeans'].value_counts()
```
